```python
#数据预处理
df['日期']= pd.to_datetime(df['日期'])  #日期时间
```

```python
def extract_values(x):
    日= x.day
    月= x.month
    年= x.year
    星期= x.day_name()
    return pd.Series([日, 月, 年, 星期])
```

```python
df[['日', '月', '年', '星期']]=df['日期'].apply(extract_values)
```

```python
df['小时'] = df['时间'].dt.hour
```

```python
fig, ax= plt.subplots(2, 2, figsize= (8,5))

ax[0, 0].plot(df[df['分店'] == 'A'].groupby('小时')['数量'].sum(), color='red')
ax[0, 1].plot(df[df['分店'] == 'B'].groupby('小时')['数量'].sum(), color='blue')
ax[1, 0].plot(df[df['分店'] == 'C'].groupby('小时')['数量'].sum(), color='yellow')

ax[0,0].set_title('分店 A')
ax[0,0].set_xlabel('小时')
ax[0,0].grid(True)
ax[0,1].set_title('分店 B')
ax[0,1].set_xlabel('小时')
ax[0,1].grid(True)
ax[1,0].set_title('分店 C')
ax[1,0].set_xlabel('小时')
ax[1,0].grid(True)

plt.subplots_adjust(hspace=0.6)


ax[1,1].axis('off');
#单独来看A分店显示其销售主要集中在14：00至16：00之间
```

```python
#让我们来看看哪个产品线的销售率最高
grouped_product= df.groupby('产品线')['评分'].mean().reset_index(drop= False)
grouped_product['评分'] = grouped_product['评分'].round(2)
grouped_product= grouped_product.sort_values(by= '评分', ascending=False)
grouped_product
```

```python
plt.figure(figsize= (10,5))
pal = 'magma'
ax = sns.barplot(y = '产品线' , x = '评分' , data = grouped_product, palette=pal)
ax.bar_label(ax.containers[0], color= 'white', padding=2)
plt.title('产品线的平均评级')
#食品和饮料的销售率最高
```

```python
#让我们看看哪个月我们的产品销售量最大
grouped_month= df.groupby('月')['数量'].sum().reset_index(drop=False)
grouped_month
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(x = '月' , y = '数量' , data = grouped_month , marker ='o')
xpoints =  grouped_month['月']
ypoints = grouped_month['数量']
for x, y in zip(xpoints, ypoints):
    plt.text(x, y, str(y), ha='center', va='bottom', color= 'white')
plt.title('按月列出的数量线图');
#我们可以说第二个月是销售量最低的月份
```
